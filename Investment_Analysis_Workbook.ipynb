```python
# Loading libraries
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing companies.csv to dataframe companies
companies = pd.read_csv('companies.csv', engine = 'python')
companies.head()
```

```python
# Importing rounds.csv to dataframe rounds
rounds2 = pd.read_csv('rounds2.csv', engine = 'python')
rounds2.head()
```

```python
# Unique Companies Present in Comapnies dataframe
companies['name'].describe()
```

```python
# Unique Companies Present in rounds2 dataframe
rounds2['company_permalink'].describe()
```

We observe that in companies dataframe, permalink has same names with lower and upper case.

We will convert all these into lower case and find unique values

```python
# Converting all permalink values of companies Dataframe to lower case
companies['permalink'] = companies['permalink'].apply(lambda x : x.lower())
companies.head()
```

```python
#Converting all comapny_permalink values of rounds2 Dataframe to lower case
rounds2['company_permalink'] = rounds2['company_permalink'].apply(lambda x: x.lower())
rounds2.head()
```

```python
# Unique values in companies dataframe
companies.permalink.describe()
```

```python
#Unique values in rounds2 dataframe
rounds2['company_permalink'].describe()
```

```python
# Percentage of null values in raised_amount_usd in rounds2 dataframe
(rounds2['raised_amount_usd'].isnull().sum()/len(rounds2.index))*100
```

```python
#Number of null values in columns
rounds2.isnull().sum()
```

```python
rounds2['raised_amount_usd'].median()
```

```python
rounds2['raised_amount_usd'].mean()
```

Since the percentage of null values is very less, we will impute them with either mean or median values

We find that the mean is very high, therefore, we impute the null values of raised_amount_usd with median values

```python
rounds2['raised_amount_usd'] = rounds2['raised_amount_usd'].fillna(rounds2['raised_amount_usd'].median())
```

```python
# Checking Number of null values
rounds2.isnull().sum()
```

```python
# Confirming data type of raised_amount_usd
rounds2['raised_amount_usd'].dtype
```

### Finding if companies in rounds2 is not present in companies dataframe

```python
rounds2['company_permalink'].isin(companies['permalink'])
```

```python
# Merging the two data frames
master_frame = pd.merge(companies, rounds2, how = 'inner', left_on = 'permalink', right_on = 'company_permalink')
master_frame.head()
```

```python
# Removing Redundant columns
master_frame = master_frame.drop(['company_permalink'], axis = 1)
```

```python
# Checking status as closed and removing the companies
master_frame = master_frame[~(master_frame['status'] == 'closed')]
```

```python
# Checking the observations in master_frame dataframe
master_frame.shape
```

### Representative amount of sectors type

###### Checkpoint 2 : Funding Type Analysis

```python
# Changing the values of raised_amount_usd to per million for ease of reading
master_frame['raised_amount_usd_per_million'] = master_frame['raised_amount_usd']/1000000
```

```python
# Creating a pivot table for Investment amount in different sectors
sectors = pd.pivot_table(data = master_frame, index = 'funding_round_type', values = 'raised_amount_usd_per_million', aggfunc = np.median)
print(sectors)
```

From the table we observe that there are two sectors which has investment between 5-15 million per round, i.e, post_ipo_equity & private_equity

The risk associated with IPOs is high because the company listed is new and the future is always at stake comapred to private equity sector, where it contains all established large organizations.

So we will keep only the companies which belong to private_equity & drop the other sectors for our further analysis

```python
# Keeping only the post_ipo_equity_values
master_frame = master_frame[(master_frame['funding_round_type'] == 'private_equity')]
master_frame.head()
```

### Checkpoint 6 : Plots

```python
# Creating master_frame1 dataframe by merging companies and rounds2 dataframes for creating plots
master_frame1 = pd.merge(companies, rounds2, how = 'inner', left_on = 'permalink', right_on = 'company_permalink')
master_frame1.head()

```
